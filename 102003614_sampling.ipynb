```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler

# load the imbalanced dataset
df = pd.read_csv('Creditcard_data.csv')


k = df.drop('Class', axis=1)
y = df['Class']
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(k),columns=k.columns)



```

**SMOTE**

```python
from imblearn.over_sampling import SMOTE, ADASYN




smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)



balanced_smote = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote)], axis=1)

balanced_smote
```

**SAMPLE SIZE DETECTION**

```python
import pandas as pd
import math

# Load the credit card dataset


# Calculate sample size
p = 0.5
E = 0.05
z = 1.96
N =len(balanced_smote)
n=math.ceil((z**2 * p * (1-p)) / (E**2 ))
print("The required sample size is:", n)
```

**CLUSTER SAMPLING**

```python
from imblearn.over_sampling._smote import cluster
import numpy as np
# Split the features and target variable
X = balanced_smote.drop('Class', axis=1)
y = balanced_smote['Class']

# Define the number of clusters
k = 385

# Create an array of cluster labels
cluster_labels = np.random.randint(0, k, size=len(X))

# Select a random sample of clusters
selected_clusters = np.random.choice(range(k), size=int(k*0.5), replace=False)

# Filter the data to include only the selected clusters
cluster_data = balanced_smote[np.isin(cluster_labels, selected_clusters)].reset_index(drop=True)
#cluster_data
len(cluster_data)
cluster_data
```

**RANDOM SAMPLING**

```python

# Split the features and target variable
X = balanced_smote.drop('Class', axis=1)
y = balanced_smote['Class']

# Perform simple random sampling
n = 385 # number of samples to select
random_sample = balanced_smote.sample(n, random_state=42)

# Split the features and target variable for the random sample
X_rs = random_sample.drop('Class', axis=1)
y_rs = random_sample['Class']
random_sample
#len(random_sample)
# X_rs
# y_rs
```

**SYSTEMATIC SAMPLING**

```python
p = 0.5
E = 0.06
z = 1.96
N =len(balanced_smote)
n=math.ceil((z**2 * p * (1-p)) / (E**2 ))

start = 0 # starting index of the sample
step = len(balanced_smote) // n # step size

systematic_sample =balanced_smote.iloc[start::step].head(n)

# Split the features and target variable for the systematic sample
X_ss = systematic_sample.drop('Class', axis=1)
y_ss = systematic_sample['Class']
systematic_sample
```

**SRATIFIED SAMPLING**

```python

import math
import random
n=math.ceil((1.96**2 * (0.25))/((0.2/2)**2))
stra_sampling=balanced_smote.groupby('Class', group_keys=False).apply(lambda x: x.sample(n))
stra_sampling
```

**CONVENIENCE**

```python
convenience_sample=balanced_smote[balanced_smote['Amount']>0.01]
convenience_sample
```

**SNOWBALL**

```python
df_snowball = balanced_smote[balanced_smote['Class'] == 1].sample(n=200, random_state=42)

# add subsequent participants based on referrals from the initial participants
for i in range(5):
    referrals = balanced_smote[balanced_smote['Class'] == 0].sample(n=40, random_state=42)
    df_snowball = pd.concat([df_snowball, referrals])
    df_snowball = df_snowball.drop_duplicates()

df_snowball
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
samples=[cluster_data,random_sample,systematic_sample,stra_sampling,df_snowball]
f_accuracy=[]
for i in samples:
  acc=[]
  X_train, X_test, y_train, y_test = train_test_split( 
   i.drop('Class',axis=1), i['Class'], test_size = 0.25, random_state = 42)
  #DecisionTree
  clf=DecisionTreeClassifier()
  clf.fit(X_train,y_train)
  y_pred=clf.predict(X_test)
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  #SVM
  classifier = SVC(kernel = 'rbf', random_state = 0)
  classifier.fit(X_train, y_train)
  y_pred=classifier.predict(X_test)
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  #Naive_Bayes
  gnb = GaussianNB()
  gnb.fit(X_train, y_train)
  y_pred = gnb.predict(X_test)
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  #KNN
  knn = KNeighborsClassifier()
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  #RandomForest
  clf = RandomForestClassifier(n_estimators = 100)  
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  accuracy=accuracy_score(y_test,y_pred)
  acc.append(accuracy)
  f_accuracy.append(acc)




  
```
